## Step 1: obtain gene sequences for all genes in the GRCm38 genome from ENSEMBL

```python
gtf_file_path = '../data/mouse/genome_files/Mus_musculus.GRCm39.113.gtf'

# Open the GTF file in read mode
with open(gtf_file_path, 'r') as gtf_file:
    # Skip the first 5 lines
    for _ in range(5):
        gtf_file.readline()

    # Initialize an empty list to store the data
    data = []
    
    # Read and process the remaining lines
    for line in gtf_file:
        # Split the line into columns
        columns = line.strip().split('\t')
        chromosome = columns[0]
        start = int(columns[3])
        stop = int(columns[4])
        category = columns[2]

        match = re.search(r'gene_name "([^"]+)"', columns[8])
        gene_name = match.group(1) if match else None

        # Append the data to the list
        data.append([chromosome, start, stop, gene_name, category])

# Create a DataFrame from the collected data
gtf_df = pd.DataFrame(data, columns=['chromosome', 'start', 'stop', 'gene_name', 'category'])

gtf_df
```

```python
gtf_df.chromosome.unique()
```

```python
valid_chromosomes = [
    '1', '2', 'X', '3', '4', '5', '6', '7', '10', '8', '14', '9', '11',
       '13', '12', '15', '16', '17', 'Y', '18', '19', 'MT'
]
gtf_df = gtf_df[gtf_df['chromosome'].isin(valid_chromosomes)]
gtf_df
```

```python
gtf_df = gtf_df[gtf_df.category=="gene"]
gtf_df
```

### Use fasta file to find sequences from gtf gene locations


```python
# Specify the path to your FASTA file containing genomic DNA
fasta_file_path = '../data/mouse/genome_files/Mus_musculus.GRCm39.dna.toplevel.fa'

# Create a dictionary to store chromosome sequences
chromosome_sequences = {}

# Iterate over the FASTA file and populate the dictionary
with open(fasta_file_path, 'r') as fasta_file:
    for record in SeqIO.parse(fasta_file, 'fasta'):
        # Keep the chromosome name as is
        chromosome_id = record.id

        chromosome_sequences[chromosome_id] = str(record.seq)

# Print available chromosome IDs from the genomic data FASTA file
#print("Chromosome IDs in FASTA file:", list(chromosome_sequences.keys()))

# Add a new column 'sequence' to the existing DataFrame
gtf_df['sequence'] = ''

# Iterate over the DataFrame rows
for index, row in gtf_df.iterrows():
    # Extract relevant information from the DataFrame
    chromosome = row['chromosome']
    start = row['start']
    stop = row['stop']

    # Use the dictionary to get the chromosome sequence
    chromosome_sequence = chromosome_sequences.get(chromosome)

    if chromosome_sequence:
        # Extract the DNA sequence based on start and stop positions
        gene_sequence = chromosome_sequence[start - 1:stop]

        # Update the 'sequence' column in the DataFrame
        gtf_df.at[index, 'sequence'] = gene_sequence

gtf_df
```

```python
gtf_df['chromosome'] = 'chr' + gtf_df['chromosome'].astype(str)
gtf_df
```

```python
gtf_df['gene_name'] = gtf_df['gene_name'].str.upper()
```

```python
gtf_df
```

```python
gtf_df = gtf_df.rename(columns={"gene_name": "gene_id"})
```

```python
gtf_df_cleaned = gtf_df.dropna(subset=['gene_id'])
gtf_df_cleaned
```

### STRING and TRRUST:

```python
mouse_tfs = pd.read_csv("../data/mouse/mouse_TF_info_scores_with_DBID.tsv", sep = '\t', index_col=0)
mouse_tfs['DBID'] = mouse_tfs['DBID'].str.upper()
mouse_tfs
```

```python
mouse_non_specific = pd.read_csv("../data/BEELINE/BEELINE-Networks/Networks/mouse/Non-Specific-ChIP-seq-network.csv", sep = ",")
mouse_STRING = pd.read_csv("../data/BEELINE/BEELINE-Networks/Networks/mouse/STRING-network.csv", sep = ",")
```

```python
mouse_ref_network = pd.concat([mouse_non_specific, mouse_STRING], axis = 0)
```

```python
print(f"Total number of unique TFs in the ref network: {len(mouse_ref_network.Gene1.unique())}")
print(f"Total number of unique genes in the ref network: {len(mouse_ref_network.Gene2.unique())}")
```

```python
# Extract the sets of valid gene and TF names
valid_genes = set(filtered_mouse_genes['gene_id'])
valid_tfs = set(mouse_tfs['DBID'])

# Filter the human_ref_network dataframe
mouse_filtered_network_2 = mouse_ref_network[
    (mouse_ref_network['Gene1'].isin(valid_tfs)) & 
    (mouse_ref_network['Gene2'].isin(valid_genes))
]

# Display the filtered dataframe
mouse_filtered_network_2
```

```python
print(f"Number of genes: {len(mouse_filtered_network_2.Gene2.unique())}")
print(f"Number of TFs: {len(mouse_filtered_network_2.Gene1.unique())}")
```

```python
# Get unique gene and TF names for rows and columns
gene_names = sorted(mouse_filtered_network_2['Gene2'].unique())
tf_names = sorted(mouse_filtered_network_2['Gene1'].unique())

# Initialize the edge matrix with zeros
mouse_edge_matrix = pd.DataFrame(0, index=gene_names, columns=tf_names)

# Fill the matrix with 1s where interactions exist
for _, row in mouse_filtered_network_2.iterrows():
    mouse_edge_matrix.at[row['Gene2'], row['Gene1']] = 1

# Display the edge matrix
mouse_edge_matrix
```

```python
print(f"Number of interactions in mouse edge matrix: {(mouse_edge_matrix == 1).sum().sum()}")
print(f"Number of non-interactions in mouse edge matrix: {(mouse_edge_matrix == 0).sum().sum()}")
```

```python
mouse_edge_matrix.to_csv("../data/mouse/Mouse_prior_knowledge_STRING_and_TRRUST.tsv", sep = '\t')
```
