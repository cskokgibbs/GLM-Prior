## Step 1: obtain gene sequences for all genes in the GRCh38 genome from ENSEMBL

```python
gtf_file_path = '../data/human/genome_files/Homo_sapiens.GRCh38.113.gtf'

# Open the GTF file in read mode
with open(gtf_file_path, 'r') as gtf_file:
    # Skip the first 5 lines
    for _ in range(5):
        gtf_file.readline()

    # Initialize an empty list to store the data
    data = []
    
    # Read and process the remaining lines
    for line in gtf_file:
        # Split the line into columns
        columns = line.strip().split('\t')
        chromosome = columns[0]
        start = int(columns[3])
        stop = int(columns[4])
        category = columns[2]

        match = re.search(r'gene_name "([^"]+)"', columns[8])
        gene_name = match.group(1) if match else None

        # Append the data to the list
        data.append([chromosome, start, stop, gene_name, category])

# Create a DataFrame from the collected data
gtf_df = pd.DataFrame(data, columns=['chromosome', 'start', 'stop', 'gene_name', 'category'])
```

```python
gtf_df
```

```python
gtf_df.chromosome.unique()
```

```python
valid_chromosomes = [
    '1', '2', '3', '4', '5', '6', '7', 'X', '8', '9', '11', '10', '12',
    '13', '14', '15', '16', '17', '18', '20', '19', 'Y', '22', '21', 'MT'
]
```

```python
gtf_df = gtf_df[gtf_df['chromosome'].isin(valid_chromosomes)]
```

```python
gtf_df = gtf_df[gtf_df.category=="gene"]
```

```python
# Specify the path to your FASTA file containing genomic DNA
fasta_file_path = '../data/human/genome_files/Homo_sapiens.GRCh38.dna.toplevel.fa'

# Create a dictionary to store chromosome sequences
chromosome_sequences = {}

# Iterate over the FASTA file and populate the dictionary
with open(fasta_file_path, 'r') as fasta_file:
    for record in SeqIO.parse(fasta_file, 'fasta'):
        # Keep the chromosome name as is
        chromosome_id = record.id

        chromosome_sequences[chromosome_id] = str(record.seq)

# Print available chromosome IDs from the genomic data FASTA file
#print("Chromosome IDs in FASTA file:", list(chromosome_sequences.keys()))

# Add a new column 'sequence' to the existing DataFrame
gtf_df['sequence'] = ''

# Iterate over the DataFrame rows
for index, row in gtf_df.iterrows():
    # Extract relevant information from the DataFrame
    chromosome = row['chromosome']
    start = row['start']
    stop = row['stop']

    # Use the dictionary to get the chromosome sequence
    chromosome_sequence = chromosome_sequences.get(chromosome)

    if chromosome_sequence:
        # Extract the DNA sequence based on start and stop positions
        gene_sequence = chromosome_sequence[start - 1:stop]

        # Update the 'sequence' column in the DataFrame
        gtf_df.at[index, 'sequence'] = gene_sequence
```

```python
sequence_lengths = human_genes['sequence'].str.len()

# Plot distribution
plt.figure(figsize=(8, 6))
plt.hist(sequence_lengths, bins=100, edgecolor='black')
plt.axvline(12000, color='red', linestyle='--', label='12kb threshold')
plt.title('Distribution of Sequence Lengths')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate sequence lengths
sequence_lengths = human_genes['sequence'].str.len()

# Filter for sequences <= 12kb
filtered_lengths = sequence_lengths[sequence_lengths <= 12000]

# Plot distribution
plt.figure(figsize=(8, 6))
plt.hist(filtered_lengths, bins=100, edgecolor='black')
plt.axvline(12000, color='red', linestyle='--', label='12kb threshold')
plt.title('Distribution of Sequence Lengths ≤ 12kb')
plt.xlabel('Sequence Length (bp)')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x')
plt.legend()
plt.show()

# Count sequences
num_less_equal_12kb = (sequence_lengths <= 12000).sum()
num_greater_12kb = (sequence_lengths > 12000).sum()

print(f"Number of sequences ≤ 12kb: {num_less_equal_12kb}")
print(f"Number of sequences > 12kb: {num_greater_12kb}")
```

## Create priors from BEELINE's datasets 

### STRING and TRRUST:

```python
human_tfs = pd.read_csv("../data/human/human_TF_info_scores_with_DBID.tsv", sep = '\t', index_col=0)
human_tfs['DBID'] = human_tfs['DBID'].str.upper()
human_tfs
```

```python
human_non_specific = pd.read_csv("../data/BEELINE/BEELINE-Networks/Networks/human/Non-Specific-ChIP-seq-network.csv", sep = ",")
human_STRING = pd.read_csv("../data/BEELINE/BEELINE-Networks/Networks/human/STRING-network.csv", sep = ",")
```

```python
human_ref_network = pd.concat([human_non_specific, human_STRING], axis = 0)
```

```python
print(f"Total number of unique TFs in the ref network: {len(human_ref_network.Gene1.unique())}")
print(f"Total number of unique genes in the ref network: {len(human_ref_network.Gene2.unique())}")
```

```python
# Extract the sets of valid gene and TF names
valid_genes = set(filtered_human_genes['gene_id'])
valid_tfs = set(human_tfs['DBID'])

# Filter the human_ref_network dataframe
human_filtered_network_2 = human_ref_network[
    (human_ref_network['Gene1'].isin(valid_tfs)) & 
    (human_ref_network['Gene2'].isin(valid_genes))
]

# Display the filtered dataframe
human_filtered_network_2
```

```python
print(f"Number of genes: {len(human_filtered_network_2.Gene2.unique())}")
print(f"Number of TFs: {len(human_filtered_network_2.Gene1.unique())}")
```

```python
# Get unique gene and TF names for rows and columns
gene_names = sorted(human_filtered_network_2['Gene2'].unique())
tf_names = sorted(human_filtered_network_2['Gene1'].unique())

# Initialize the edge matrix with zeros
human_edge_matrix = pd.DataFrame(0, index=gene_names, columns=tf_names)

# Fill the matrix with 1s where interactions exist
for _, row in human_filtered_network_2.iterrows():
    human_edge_matrix.at[row['Gene2'], row['Gene1']] = 1

# Display the edge matrix
human_edge_matrix
```

```python
print(f"Number of interactions in human edge matrix: {(human_edge_matrix == 1).sum().sum()}")
print(f"Number of non-interactions in human edge matrix: {(human_edge_matrix == 0).sum().sum()}")
```

```python
human_edge_matrix.to_csv("../data/human/Human_prior_knowledge_STRING_and_TRRUST.tsv", sep = '\t')
```
